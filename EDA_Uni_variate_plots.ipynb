# **Uni variate Plots**

## **Displot (Histogram)**

```python
sns.displot(df1.rent,bins=50,kde=False)
```

## **Box plot**

```python
sns.boxplot(x='rent',data=df1)
plt.show()
```

## **Pie Chart (Malplotlib)**

```python
dt1=df1.iloc[0:10,:]  # seperate first 10 datas
x=dt1.bedrooms        # X Data = Bed Room Count
my_label=dt1.rent     # Label Data = Rent
plt.pie(x,labels=my_label)
plt.show()
```

## **Categorical scatterplots**

```python
sns.catplot(data=df1, x="rent", hue="neighborhood", kind="swarm")
plt.show()
```

```python
sns.catplot(data=df1, x="rent", kind="violin")
plt.show()
```

## **Scatterplots (Matplotlib)**

```python
plt.scatter(dt1.rent,dt1.bedrooms)
plt.title('Rent vs bedrooms')
plt.xlabel('Rent')
plt.ylabel('Bedrooms')
```
